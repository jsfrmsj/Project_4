```python
sleep_dummies_smoking = pd.get_dummies(sleep_df_2["Smoking status"])

# Concatenate the "Smoking status" variables with the DataFrame.
sleep_df_clean1 = pd.concat([sleep_df_clean1, sleep_dummies_smoking], axis=1)

# Display the sample data
sleep_df_clean1.head()
```

```python
sleep_dummies_wakeup = pd.get_dummies(sleep_df_2["Wakeup Time"])

# Concatenate the "wakeup time" variables with the DataFrame.
sleep_df_clean1 = pd.concat([sleep_df_clean1, sleep_dummies_wakeup], axis=1)

# Display the sample data
sleep_df_clean1.head()
```

```python
sleep_dummies_efficiency = pd.get_dummies(sleep_df_2["Sleep efficiency"])

# Concatenate the "wakeup time" variables with the DataFrame.
sleep_df_clean1 = pd.concat([sleep_df_clean1, sleep_dummies_efficiency], axis=1)

# Display the sample data
sleep_df_clean1.head()
```

```python
#correlations_encoded = sleep_df_clean1.corr()
#correlations_encoded["Sleep efficiency"]
```

```python
sleep_df_cleandrop = sleep_df_clean1.drop(columns= ["ID", "Sleep duration", "REM sleep percentage",
                                                    "Deep sleep percentage", "Light sleep percentage",
                                                   "Awakenings", "Gender", "Bedtime Hour",
                                                    "Wakeup Time", "Smoking status", "Sleep efficiency", "Inefficient"])

sleep_df_cleandrop.head()
```

```python
# Seperate the features, X,  from the target variable, y
y = sleep_df_cleandrop['Efficient']
X = sleep_df_cleandrop.drop(columns='Efficient')
```

```python
# Split the dataset using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
 X_train.shape
```

```python
# Create the StandardScaler instance
scaler = StandardScaler()
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)
# Scale the training data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Instantiate the model with k = 3 neighbors
model = KNeighborsClassifier(n_neighbors=3)
```

```python
 # Train the model
model.fit(X_train_scaled, y_train)
```

```python
KNeighborsClassifier(n_neighbors=3)
```

```python
# Create predictions
y_pred = model.predict(X_test_scaled)

# Review the predictions
y_pred
```

```python
# Print confusion matrix
confusion_matrix(y_pred,y_test)
```

```python
# Print classification report
print(classification_report(y_pred,y_test))
```
