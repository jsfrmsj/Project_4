```python
import pandas as pd
import numpy as np
```

```python
# Loading dataset 1
sleep_df = pd.read_csv("new_sleep_data.csv")

sleep_df.head()
```

```python
# Loading dataset 2
sleep_df_2 = pd.read_csv("Sleep_Dataset_2.csv")

sleep_df_2.head()
```

```python
#merge datasets
frames = [sleep_df, sleep_df_2]
sleep_df_merged = pd.concat(frames)

sleep_df_merged.head()
```

```python
sleep_df_merged.tail()
```

```python
# Get Dummies for Gender Variable
sleep_dummies_gender = pd.get_dummies(sleep_df_merged["Gender"])

# Concatenate the "gender" variables with the DataFrame.
sleep_df_dummies = pd.concat([sleep_df_merged, sleep_dummies_gender], axis=1)

# Display the sample data
sleep_df_dummies.head()
```

```python
# Get dummies for Bedtime Hour variable
sleep_dummies_bedtime = pd.get_dummies(sleep_df["Bedtime Hour"])

# Concatenate the "bedtime hours" variables with the DataFrame.
sleep_df_dummies = pd.concat([sleep_df_dummies, sleep_dummies_bedtime], axis=1)

# Display the sample data
sleep_df_dummies.head()
```

```python
# Get dummies for Smoking status variable
sleep_dummies_smoking = pd.get_dummies(sleep_df["Smoking status"])

# Concatenate the "Smoking status" variables with the DataFrame.
sleep_df_dummies = pd.concat([sleep_df_dummies, sleep_dummies_smoking], axis=1)

# Display the sample data
sleep_df_dummies.head()
```

```python
# Get dummies for Wakeup Time variable
sleep_dummies_wakeup = pd.get_dummies(sleep_df["Wakeup Time"])

# Concatenate the "wakeup time" variables with the DataFrame.
sleep_df_dummies = pd.concat([sleep_df_dummies, sleep_dummies_wakeup], axis=1)

# Display the sample data
sleep_df_dummies.head()
```

```python
sleep_df_dummies.to_csv("sleep_df_dummies.csv")
```

```python
# Get dummies for the Target Variable - Sleep Efficiency
sleep_dummies_efficiency = pd.get_dummies(sleep_df["Sleep efficiency"])

# Concatenate the "wakeup time" variables with the DataFrame.
sleep_df_dummies = pd.concat([sleep_df_dummies, sleep_dummies_efficiency], axis=1)

# Display the sample data
sleep_df_dummies.head()
```

```python
# Drop redundant and not needed columns
sleep_df_clean = sleep_df_dummies.drop(columns= ["ID", "REM sleep percentage",
                                                    "Deep sleep percentage", "Light sleep percentage",
                                                   "Awakenings", "Gender", "Bedtime Hour",
                                                    "Wakeup Time", "Smoking status", "Sleep efficiency",
                                                    "Inefficient", "Age"])

sleep_df_clean.head()
```
