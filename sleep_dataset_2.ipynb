```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

# Read the sleep dataset 2 and the study results
dataset_2 = pd.read_csv("C:\\Users\\dtrob\\repos\\Project-4\\Project_4\\PLACES_Place_Data2021.csv")
dataset_2
```

```python
# Create a dictionary to map the old column names to the new column names
rename_sleep_data = {
    'StateAbbr': 'State',
    'TotalPopulation': 'Population',
    'BINGE_CrudePrev': 'Alcohol Consumption',
    'CSMOKING_CrudePrev': 'Smoking Habit'
}

# Use the rename() method with the rename_sleep_data dictionary
dataset_2 = dataset_2.rename(columns=rename_sleep_data)
```

```python
# Specify the columns you want to keep
columns_to_keep = ['State', 'Population', 'Alcohol Consumption', 'Smoking Habit']

# Use the drop() method to drop the columns not in columns_to_keep
dataset_2 = dataset_2.drop(columns=set(dataset_2.columns) - set(columns_to_keep))
dataset_2
```

```python
# Remove rows with any missing values
dataset_2 = dataset_2.dropna()
dataset_2
```

```python
# Group the data by 'State' and calculate the mean for each state
state_averages = dataset_2.groupby('State').mean()

# Reset the index to make the 'State Abbreviation' column a regular column
state_averages = state_averages.reset_index()
state_averages
```

```python
# Specify the file path and name for the CSV file
file_path = "C:\\Users\\dtrob\\repos\\Project-4\\Project_4\\state_averages.csv"

# Export the DataFrame to a CSV file
state_averages.to_csv(file_path, index=False)
```
